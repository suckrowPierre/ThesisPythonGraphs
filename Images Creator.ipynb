```python
from PIL import Image

img_folder = 'img/'


def combine_instruments4Block( title ):
    if title == 'Weißes Rauschen':
        img1 = Image.open(f"{img_folder}{title}.png")
        img2 = Image.open(f"{img_folder}Spektrum von weißem Rauschen.png")
        img3 = Image.open(f"{img_folder}Spektrogramm von weißem Rauschen.png")
        img4 = Image.open(f"{img_folder}Mel Spektrogramm von weißem Rauschen.png")
    else:
        img1 = Image.open(f"{img_folder}{title}.png")
        img2 = Image.open(f"{img_folder}Spektrum einer {title}.png")
        img3 = Image.open(f"{img_folder}Spektrogramm einer {title}.png")
        img4 = Image.open(f"{img_folder}Mel Spektrogramm einer {title}.png")

    width, height = img1.size

    combined = Image.new('RGB', (width * 2, height * 2))

    combined.paste(img1, (0, 0))
    combined.paste(img2, (width, 0))
    combined.paste(img3, (0, height))
    combined.paste(img4, (width, height))

    combined.save(f"{img_folder}combined{title}.png")





combine_instruments4Block('Gitarrenseite')
combine_instruments4Block('Klavierseite')
combine_instruments4Block('Sinusschwingung')
combine_instruments4Block('Rechteckschwingung zusammengesetzt aus 50 harmonsichen Schwingungen')
combine_instruments4Block('Sägezahnschwingung zusammengesetzt aus 50 harmonsichen Schwingungen')
combine_instruments4Block('Weißes Rauschen')
```

```python
def evolution():
    harmonics = [2,5,50]
    size = len(harmonics)
    saw = []
    square = []
    
    # Define new width and height
    img_harmonics = Image.open(f"{img_folder}harmonische Obertöne einer Sinusschwingung.png")
    width, height = img_harmonics.size
    new_width = 1000
    new_height = int(new_width * height / width)
    

    for x in harmonics:
        img = Image.open(f"{img_folder}Sägezahnschwingung zusammengesetzt aus {x} harmonsichen Schwingungen.png")
        resized_img = img.resize((new_width, new_height))
        saw.append(resized_img)  # add to saw image array 
        
    harmonics = [3,50]
    square.append(img_harmonics.resize((new_width, new_height)))
    
    for x in harmonics:
        img = Image.open(f"{img_folder}Rechteckschwingung zusammengesetzt aus {x} harmonsichen Schwingungen.png")
        resized_img = img.resize((new_width, new_height))
        square.append(resized_img)  # add to square image array 

    combined = Image.new('RGB', (new_width * 2, new_height * size))
    
    for x in range(size):
        combined.paste(square[x], (0, new_height * x))
        combined.paste(saw[x], (new_width, new_height * x))
    
    combined.save(f"{img_folder}evolution.png")

evolution()
```

```python
def combine_spectograms():
    imgs = [ Image.open(f"{img_folder}combinedSinusschwingung.png"), 
            Image.open(f"{img_folder}combinedRechteckschwingung zusammengesetzt aus 50 harmonsichen Schwingungen.png"),
            Image.open(f"{img_folder}combinedSägezahnschwingung zusammengesetzt aus 50 harmonsichen Schwingungen.png"),
            Image.open(f"{img_folder}combinedKlavierseite.png"),
            Image.open(f"{img_folder}combinedGitarrenseite.png"),
            Image.open(f"{img_folder}combinedWeißes Rauschen.png")
           ]
    

    width, height = imgs[0].size

    combined = Image.new('RGB', (width, height * 6))

    
    for x in range(6):
        combined.paste(imgs[x], (0, height * x))

    combined.save(f"{img_folder}spectrums.png")

combine_spectograms()

def combine_spectograms2():
    imgs = [ Image.open(f"{img_folder}combinedSinusschwingung.png"), 
            Image.open(f"{img_folder}combinedRechteckschwingung zusammengesetzt aus 50 harmonsichen Schwingungen.png"),
            Image.open(f"{img_folder}combinedSägezahnschwingung zusammengesetzt aus 50 harmonsichen Schwingungen.png"),
            Image.open(f"{img_folder}combinedKlavierseite.png"),
            Image.open(f"{img_folder}combinedGitarrenseite.png"),
            Image.open(f"{img_folder}combinedWeißes Rauschen.png")
           ]
    

    width, height = imgs[0].size

    combined = Image.new('RGB', (width*2, height * 3))

    
    i = 0
    b = 0
    while i < 6:
        combined.paste(imgs[i], (0, height * b))
        i += 1
        combined.paste(imgs[i], (width, height * b))
        i += 1
        b += 1


    combined.save(f"{img_folder}spectrums2.png")

combine_spectograms2()
```

```python
def combine_instruments3( title ):
    if title == 'Weißes Rauschen':
        img1 = Image.open(f"{img_folder}Spektrum von weißem Rauschen.png")
        img2 = Image.open(f"{img_folder}Spektrogramm von weißem Rauschen.png")
        img3 = Image.open(f"{img_folder}Mel Spektrogramm von weißem Rauschen.png")
    else:
        img1 = Image.open(f"{img_folder}Spektrum einer {title}.png")
        img2 = Image.open(f"{img_folder}Spektrogramm einer {title}.png")
        img3 = Image.open(f"{img_folder}Mel Spektrogramm einer {title}.png")

    width, height = img1.size

    combined = Image.new('RGB', (width * 3, height))

    combined.paste(img1, (0, 0))
    combined.paste(img2, (width, 0))
    combined.paste(img3, (width*2, 0))

    combined.save(f"{img_folder}combined_flat{title}.png")
    
combine_instruments3('Gitarrenseite')
combine_instruments3('Klavierseite')
combine_instruments3('Sinusschwingung')
combine_instruments3('Rechteckschwingung zusammengesetzt aus 50 harmonsichen Schwingungen')
combine_instruments3('Sägezahnschwingung zusammengesetzt aus 50 harmonsichen Schwingungen')
combine_instruments3('Weißes Rauschen')
```
