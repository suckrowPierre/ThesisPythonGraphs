```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

import numpy as np

def generate_waveform(freq, sample_rate, duration, wave_type, num_harmonics):
    t = np.linspace(0., duration, int(sample_rate * duration), False)  # Time array

    if wave_type == 'sinus':
        waveform = np.sin(2. * np.pi * freq * t)
    elif wave_type == 'square':
        waveform = np.zeros_like(t)
        for h in range(1, num_harmonics+1, 2):
            waveform += np.sin(2. * np.pi * h * freq * t) / h
        waveform *= (4. / np.pi)
    elif wave_type == 'sawtooth':
        waveform = np.zeros_like(t)
        for h in range(1, num_harmonics+1):
            waveform += np.sin(2. * np.pi * h * freq * t) / h
        waveform *= (-2. / np.pi)
    return waveform


def plot_waveform(sample_rate, waveform, title="", time_range_ms=10.0):
    duration = len(waveform) / sample_rate
    t = np.linspace(0., duration, len(waveform), False)

    # Plot only the first few milliseconds defined by time_range_ms
    time_range_samples = int(time_range_ms * sample_rate / 1000)
    
    plt.figure(figsize=(10, 4))
    plt.plot(t[:time_range_samples], waveform[:time_range_samples])
    plt.title(title)
    plt.grid()
    plt.xlabel ('Zeit')
    plt.ylabel ('Amplitude')
    plt.show()

def plot_harmonics(freq, sample_rate, duration, num_harmonics, title=""):
    t = np.linspace(0., duration, int(sample_rate * duration), False)
    plt.figure(figsize=(10, 4))

    for h in range(1, num_harmonics + 1):
        overtone = np.sin(2. * np.pi * h * freq * t) / h
        plt.plot(t[:int(10 * sample_rate / 1000)], overtone[:int(10 * sample_rate / 1000)],
                 label='Oberton ' + str(h))

    plt.title('harmonische Obertne einer ' + title)
    plt.grid()
    plt.legend(loc='upper right')
    plt.xlabel ('Zeit')
    plt.ylabel ('Amplitude')
    plt.show()

def plot_spectrogram(waveform, sr, n_fft=2048, hop_length=512, title=""):
    plt.figure(figsize=(10, 4))
    D = librosa.amplitude_to_db(np.abs(librosa.stft(waveform, n_fft=n_fft, hop_length=hop_length)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spektrogramm einer ' + title)
    plt.tight_layout()
    plt.xlabel ('Zeit')
    plt.ylabel ('Frequenz')
    plt.show()

def plot_melspectrogram(waveform, sr, n_fft=2048, hop_length=512, n_mels=128, title=""):
    plt.figure(figsize=(10, 4))
    S = librosa.feature.melspectrogram(y=waveform, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
    S_DB = librosa.amplitude_to_db(S, ref=np.max)
    librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spektrogramm einer ' + title)
    plt.tight_layout()
    plt.xlabel ('Zeit')
    plt.ylabel ('Frequenz')
    plt.show()

def plot_spectrum(waveform, sample_rate, title=""):
    # Perform Fourier transform
    fft = np.fft.rfft(waveform)

    # Compute the frequencies corresponding to the fft bins
    freqs = np.fft.rfftfreq(len(waveform), 1 / sample_rate)

    # Compute the magnitude of the fft (absolute value)
    magnitude = np.abs(fft)

    # Plot the spectrum
    plt.figure(figsize=(10, 4))
    plt.plot(freqs, magnitude)
    plt.title('Spectrum of ' + title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid()
    plt.show()

def plot_waveform_spectrogram_spectrum(waveform, sample_rate, title=""):
    plot_waveform(sample_rate, waveform, title)
    plot_spectrogram(waveform, sample_rate, title)
    plot_melspectrogram(waveform, sample_rate, title)
    plot_spectrum(waveform, sample_rate, title)

# parameters
sample_rate = 44100.0
freq = 440.0
duration = 1.0  # in seconds
num_harmonics = 50

wave_types = ['sinus', 'saw', 'square']
harmonics_to_display = [2, 3, 5, 10, 20, 50]

for wave_type in wave_types:
    waveform = generate_waveform(freq, sample_rate, duration, wave_type, num_harmonics)
    plot_waveform_spectrogram_spectrum(waveform, sample_rate, title=wave_type)
    if wave_type == 'sinus':
        plot_harmonics(freq, sample_rate, duration, 10, title=wave_type)
    else:
        for h in harmonics_to_display:
            waveform_harmonic = generate_waveform(freq, sample_rate, duration, wave_type, h)
            plot_waveform(sample_rate, waveform_harmonic, title=f"{wave_type} zusammengesetzt aus {h} harmonischen Schwingungen")

# Load the audio files
guitar = 'BasicGuitar440Hz.wav'
piano = 'GrandPiano440Hz.wav'

# Load the waveforms
guitar_waveform, guitar_sample_rate = librosa.load(guitar, sr=44100)
piano_waveform, piano_sample_rate = librosa.load(piano, sr=44100)

plot_waveform_spectrogram_spectrum(guitar_waveform, guitar_sample_rate, title="Gitarrenseite")
plot_waveform_spectrogram_spectrum(piano_waveform, piano_sample_rate, title="Klavierseite")
```
